##**Comparing all models together**

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics names without MCC and AUROC
metrics = ["Accuracy", "F1 Score", "Recall (Sensitivity)", "Precision"]

# TF-IDF + Logistic Regression values without MCC and AUROC
tfidf_values = [
    0.8101,              # Accuracy
    0.6530,              # F1 Score
    0.8734,              # Recall
    34 / (34 + 0)        # Precision = 1.0
]

# DistilBERT values without MCC and AUROC
distilbert_values = [
    0.9641,              # Accuracy
    0.8262,              # F1 Score
    48 / (48 + 47),      # Recall â‰ˆ 0.5053
    48 / (48 + 0)        # Precision = 1.0
]

# T5 values without MCC and AUROC
t5_values = [
    0.9314,              # Accuracy
    0.5494,              # F1 Score (Macro avg)
    0.5360,              # Recall (Macro avg)
    0.7778               # Precision
]

# Bar width and X locations
x = np.arange(len(metrics))
width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, tfidf_values, width, label='LR + TF-IDF', color='lightcoral')
plt.bar(x, distilbert_values, width, label='DistilBERT', color='skyblue')
plt.bar(x + width, t5_values, width, label='T5', color='lightgreen')

# Labels and formatting
plt.ylabel('Score')
plt.title('Comparison of Evaluation Metrics: LR + TF-IDF vs. DistilBERT vs. T5')
plt.xticks(x, metrics, rotation=45)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

metrics = ["Accuracy", "F1 Score", "Recall", "Precision", "MCC", "AUROC"]
labels=np.array(metrics)

tfidf_values = [0.8101, 0.6530, 0.8734, 1.0, 0.56, 0.9407]
distilbert_values = [0.9641, 0.8262, 0.5053, 1.0, 0.6975, 0.9514]
t5_values = [0.9314, 0.5494, 0.5360, 0.7778, 0.1519, 0.799]

values = [tfidf_values, distilbert_values, t5_values]
model_names = ['LR + TF-IDF', 'DistilBERT', 'T5']

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(7,7), subplot_kw=dict(polar=True))

for idx, val in enumerate(values):
    val += val[:1]
    ax.plot(angles, val, label=model_names[idx], linewidth=2)
    ax.fill(angles, val, alpha=0.25)

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
ax.set_ylim(0,1)
plt.title('Model Performance Radar Chart')
plt.legend(loc='upper right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['LR + TF-IDF', 'DistilBERT', 'T5']
mcc_values = [0.56, 0.6975, 0.1519]
auroc_values = [0.9407, 0.9514, 0.799]

x = np.arange(len(models))

plt.figure(figsize=(8, 6))

plt.plot(x, mcc_values, marker='o', linestyle='-', color='red', label='MCC')
plt.plot(x, auroc_values, marker='s', linestyle='--', color='blue', label='AUROC')

plt.xticks(x, models)
plt.ylim(0, 1.1)
plt.xlabel('Models')
plt.ylabel('Score')
plt.title('MCC & AUROC Comparison Across Models')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
